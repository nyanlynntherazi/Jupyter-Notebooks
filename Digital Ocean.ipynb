```python
all_years = []

for year in range(1880, 2016+1):
    all_years.append(pd.read_csv('names/yob{}.txt'.format(year),names = ['Name', 'Sex', 'Babies']))
    all_years[-1]['Year'] = year

all_names = pd.concat(all_years)
```

```python
all_names_index = all_names.set_index(['Sex','Name','Year']).sort_index()
```

```python
all_names_index
```

```python
def name_plot(sex, name):
    data = all_names_index.loc[sex, name]

    plt.pyplot.plot(data.index, data.values)
```

```python
name_plot('F', 'Danica')
```
